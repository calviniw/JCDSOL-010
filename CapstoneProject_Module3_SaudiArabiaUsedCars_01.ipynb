**Metric Evaluation**


Pada model kali ini, penghapusan outlier hanya dilakukan pada data outlier dengan nilai yang ekstrem dan kurang sesuai dengan domain knowledge. Dengan demikian, sejumlah outlier masih akan terdapat dalam data. Hal ini menyebabkan metrik evaluasi model yang akan digunakan tidak begitu sensitif terhadap adanya outlier.

Oleh karena itu, beberapa metrik evaluasi yang relevan dalam model ini mencakup:

1. **MAPE (Mean Absolute Percentage Error)**
  
  Mengukur rata-rata persentase selisih absolut antara prediksi dan nilai aktual. MAPE berbeda dari MAE, karena MAPE akan mengubah selisih menjadi persentase. Dikarenakan dalam bentuk persentase, nilai metrik ini berada dalam rentang 0 hingga 1.


2. **MAE (Mean Absolute Error)**

  Mengukur selisih rata-rata absolut antara nilai prediksi dan nilai aktual. Metrik ini direkomendasikan untuk digunakan karena data masih mengandung outlier. MAE tidak memperhatikan selisih yang menghasilkan nilai negatif. Semakin kecil nilai MAE, maka semakin baik prediksi model.


3. **RMSLE (Root Mean Squared Log Error)**
  
  Direkomendasikan untuk digunakan ketika data memiliki skewness yang tinggi dengan variasi yang bervariasi dari kecil hingga besar. RMSLE mengubah data menjadi skala logaritmik, lalu menghitung rata-ratanya dan mengambil akar kuadrat. Karena sifat logaritmik, metrik ini cenderung lebih tahan terhadap perbedaan yang besar.


4. **R-Squared**

  Metrik ini mengukur seberapa besar proporsi varians dalam data target dapat dijelaskan oleh model. Semakin tinggi nilainya, semakin baik model mampu menjelaskan variasi dalam data observasi. Metrik ini lebih sesuai digunakan saat memilih model dasar karena dapat menunjukkan seberapa besar pengaruh fitur terhadap target.

Metrik lainnya seperti RMSE (Root Mean Squared Error), MSE (Mean Squared Error), RMPSE (Root Mean Square Percentage Error) kurang cocok digunakan dalam konteks ini karena sensitivitas terhadap adanya outlier. Outlier memiliki dampak signifikan pada metrik-metrik tersebut karena mereka menghitung perbedaan kuadrat atau persentase antara prediksi dan aktual. Dalam hal ini, menggunakan metrik evaluasi yang lebih stabil terhadap outlier seperti MAE merupakan pendekatan yang lebih tepat. Dengan demikian, evaluasi kualitas prediksi model akan lebih akurat dan informatif.

Pra-pemrosesan data merujuk pada langkah-langkah yang diambil sebelum data dianalisis atau dimasukkan ke dalam model machine learning. Dalam tahap analisis berikutnya, langkah yang perlu diambil adalah mengidentifikasi dan menghapus indeks yang tidak relevan berdasarkan pengetahuan di bidang tersebut. Akan tetapi, tidak semua data outlier harus dihilangkan karena kemungkinan data tersebut sesuai dengan fakta dan realitas yang ada. Namun, nilai outlier yang sangat ekstrem dapat berpengaruh besar terhadap hasil interpretasi yang kita peroleh dari data atau model machine learning yang digunakan. Dengan tindakan yang sesuai, kita dapat menganggap outlier ekstrem sebagai populasi yang memang berbeda.

Sebelum kita menelaah lebih dalam pada data outlier ekstrem, langkah awalnya adalah memeriksa indeks-indeks yang tampaknya tidak masuk akal atau tidak memberikan kontribusi yang signifikan. Dalam kerangka ini, kita mencari indeks-indeks yang tidak memiliki relevansi atau dampak yang berarti dalam analisis yang dilakukan. Dengan menghapus indeks-indeks yang tidak memberikan nilai tambah, kita dapat menyederhanakan dataset dan berfokus pada variabel-variabel yang memiliki peran penting dalam pembangunan model prediktif yang akurat.

```python
# Check outlier berdasarkan IQR (Interquartile)
def outlierChecker(col):

    Q1 = np.percentile(col, 25)
    Q3 = np.percentile(col, 75)

    IQR = Q3 - Q1
    batas_bawah = Q1 - (IQR * 1.5)
    batas_atas = Q3 + (IQR * 1.5)

    outliersData = col[(col < batas_bawah) | (col > batas_atas)]
    return outliersData
```

```python
# Summary Outlier dari Variable Numerik
fiturOutlier = []
for col in df.select_dtypes(['float','int']):
    outliers = outlierChecker(df[col])
    if len(outliers):
        print(f"Fitur",col, "terdapat", len(outliers), "outlier")
        fiturOutlier.append(col)
    else:
        print(f"Fitur", col, "tidak terdapat outlier")
```

Multicollinearity merujuk pada situasi di mana dua atau lebih fitur dalam model regresi memiliki korelasi yang tinggi antara satu sama lain. Fenomena ini dapat mengakibatkan tantangan dalam memahami hasil model, karena bisa sulit untuk memisahkan kontribusi masing-masing fitur terhadap variabel target. Dalam konteks ini, penting untuk memeriksa apakah ada korelasi kuat antara fitur-fitur lain dalam dataset yang dapat memengaruhi interpretasi model.

Untuk mengevaluasi adanya multicollinearity, kita dapat menggunakan matriks korelasi antara semua fitur yang ada dalam dataset. Jika terdapat korelasi yang tinggi antara dua fitur atau lebih, itu dapat mengindikasikan adanya multicollinearity.

Selain itu, kita juga dapat menggunakan faktor inflasi varians (VIF) untuk mengukur tingkat multicollinearity antara fitur-fitur. Jika nilai VIF suatu fitur melebihi batas yang ditentukan, hal ini menandakan adanya multicollinearity yang signifikan dengan fitur lain dalam model.

```python
# Check Variance Inflation Factor (VIF)
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    vif = pd.DataFrame()
    vif['variables'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)
calc_vif(df[fiturOutlier])
```

Dalam kerangka teori yang umum digunakan, biasanya nilai VIF (Variance Inflation Factor) yang melebihi 4 dianggap sebagai tanda adanya multicollinearity. Namun, dalam situasi ini, kami menyarankan untuk tidak menghapus fitur dengan nilai VIF lebih dari 4, seperti "Year" dan "Engine_Size", meskipun terdapat indikasi multicollinearity. Ini disebabkan oleh dampak penting fitur-fitur tersebut terhadap prediksi penjualan mobil bekas.

Dalam beberapa kasus, meskipun multicollinearity hadir, fitur-fitur tersebut dapat tetap memberikan informasi yang signifikan dalam memprediksi variabel target. Dalam konteks ini, "Year" dan "Engine_Size" mungkin memberikan wawasan penting tentang nilai mobil bekas dan preferensi konsumen terkait ukuran mesin. Karena itu, meskipun multicollinearity ada, sebaiknya fitur-fitur ini tetap dipertahankan dalam model.

Namun, perlu diingat bahwa adanya multicollinearity dapat mempengaruhi cara kita menafsirkan hasil model. Oleh karena itu, saat menganalisis hasil model, kita harus berhati-hati dalam mengartikan keterkaitan dari masing-masing fitur yang terlibat dalam multicollinearity.

**Encoding**

Untuk menggunakan semua fitur yang telah dipilih dalam pembuatan model, transformasi dari variable kategorikal ke variable numerikal diperlukan. Ini dapat dilakukan melalui penggunaan ColumnTransformer, sebuah alat dalam pemrosesan data yang memungkinkan penerapan transformasi yang berbeda pada setiap kolom data.

ColumnTransformer berperan sebagai alat yang memfasilitasi transformasi spesifik pada setiap kolom dalam dataset. Dalam kasus ini, penggunaan ColumnTransformer berguna untuk mengubah variable kategorikal menjadi bentuk numerikal. Teknik yang digunakan untuk transformasi ini bisa berupa One-Hot Encoding atau Label Encoding, tergantung pada jenis data dan konteks fitur yang sedang diolah.

Setelah ColumnTransformer didefinisikan dengan transformasi yang sesuai untuk setiap kolom, langkah selanjutnya adalah mengintegrasikannya ke dalam pipeline model. Pipeline model merupakan susunan langkah pemrosesan data yang terdiri dari serangkaian transformasi dan estimasi model. Dalam hal ini, ColumnTransformer akan menjadi salah satu komponen dalam pipeline, berfungsi untuk mengubah variable kategorikal menjadi bentuk numerikal sebelum proses pemodelan dilakukan.

Dengan menggunakan ColumnTransformer dan pipeline model, kita mampu dengan efisien mengkonversi variable kategorikal menjadi format numerikal. Ini memungkinkan kita untuk memanfaatkan seluruh fitur yang telah dipilih dalam proses pembuatan model.

Pada bagian kode tersebut, kita membuat sebuah objek ColumnTransformer dengan variable bernama `ct`. ColumnTransformer digunakan untuk menggabungkan beberapa transformer yang berbeda pada kolom-kolom yang berbeda dalam dataset yang sama.

Berikut adalah penjelasan untuk setiap argumen yang digunakan dalam ColumnTransformer:

- `'One Hot Encoder'`
  
  Label untuk transformer pertama yang akan digunakan.
- `OneHotEncoder(drop='first')`

  Menentukan perlakuan terhadap kolom-kolom yang tidak mengalami transformasi. `'passthrough'` merupakan kolom-kolom yang tidak di-encode atau di-transformasikan akan diteruskan (tanpa perubahan) ke hasil keluaran. Dengan menggunakan `ColumnTransformer`, kita dapat menggabungkan beberapa transformer yang berbeda dalam satu proses dan menerapkan transformasi ini pada dataset dengan mudah. 
**Memilih Benchmark Model yang Tepat**

```python
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=100)
rf = RandomForestRegressor(random_state=100)
gdt = GradientBoostingRegressor(random_state=100)
xgb = XGBRegressor(random_state=100)
ridge = Ridge()
lasso = Lasso()
elastic_net = ElasticNet()
Ada_Boost = AdaBoostRegressor()
lgbm = LGBMRegressor(random_state=100)
cat = CatBoostRegressor(random_state=100, verbose=False)


ttr_lr = TransformedTargetRegressor(lr, func = np.log, inverse_func = np.exp)
ttr_knn = TransformedTargetRegressor(knn, func = np.log, inverse_func = np.exp)
ttr_dt = TransformedTargetRegressor(dt, func = np.log, inverse_func = np.exp)
ttr_rf = TransformedTargetRegressor(rf, func = np.log, inverse_func = np.exp)
ttr_gdt = TransformedTargetRegressor(gdt, func = np.log, inverse_func = np.exp)
ttr_xgb = TransformedTargetRegressor(xgb, func = np.log, inverse_func = np.exp)
ttr_ridge = TransformedTargetRegressor(ridge, func = np.log, inverse_func = np.exp)
ttr_lasso = TransformedTargetRegressor(lasso, func = np.log, inverse_func = np.exp)
ttr_enet = TransformedTargetRegressor(elastic_net, func = np.log, inverse_func = np.exp)
ttr_Ada_Boost = TransformedTargetRegressor(Ada_Boost, func = np.log, inverse_func = np.exp)
ttr_lgbm = TransformedTargetRegressor(lgbm, func = np.log, inverse_func = np.exp)
ttr_cat = TransformedTargetRegressor(cat, func = np.log, inverse_func = np.exp)


models = [ttr_lr, ttr_knn, ttr_dt, ttr_rf, ttr_gdt, ttr_xgb,  ttr_ridge, ttr_lasso, ttr_enet, ttr_Ada_Boost, ttr_lgbm, ttr_cat]


score_mae = []
score_mape = []
score_rmsle = []
score_r2 = []

for model in models:

    cross_val = KFold(n_splits = 5, shuffle = True, random_state = 100)
    estimator = Pipeline([
        ('preprocessing', ct),
        ('model', model)
    ])

    # MAE
    model_cv_mae = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv = cross_val,
        scoring = 'neg_mean_absolute_error',
        error_score = 'raise'
        )

    score_mae.append(model_cv_mae.mean())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv = cross_val,
        scoring = 'neg_mean_absolute_percentage_error',
        error_score = 'raise'
        )

    score_mape.append(model_cv_mape.mean())

       # RMSLE
    model_cv_rmsle = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv = cross_val,
        scoring = 'neg_mean_squared_log_error',
        error_score = 'raise'
        )

    score_rmsle.append(model_cv_rmsle.mean())

       # R2
    model_cv_r2  =  cross_val_score(
        estimator,
        X_train,
        y_train,
        cv = cross_val,
        scoring = 'r2',
        error_score = 'raise'
        )

    score_r2.append(model_cv_r2.mean())

score_rmsle = np.sqrt(abs(np.array(score_rmsle)))
```

**Hyperparameter Tuning**

Dalam konteks ini, kami akan mencari kombinasi terbaik untuk model Cat Boost menggunakan metode grid search atau random search.
```python
hyperparam_space_catboost = {
    'model__learning_rate': [0.01, 0.1, 0.5],
    'model__iterations': [175, 225, 275],
    'model__depth': [i for i in range(7,11,1)],
    'model__l2_leaf_reg': [i for i in range(4,7,1)]
}
```
